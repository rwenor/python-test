```python
def prt_mean_std(col, txt):
    print("%s: %0.1f (+-%.0f)" % (txt, col.mean(), col.std()))
```

```python
print "\nHastighet i 0.01 km/t\n", "-"*25, "\n"
prt_mean_std(T2.HA_diff, "Piezo diff")
prt_mean_std(T2.HL_diff, "Loop diff")
prt_mean_std(T2.HLFA_diff, "Loop inn diff(piezo)")
prt_mean_std(T2.HLTA_diff, "Loop ut diff(piezo)")
```
